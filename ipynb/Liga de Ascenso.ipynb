## Separamos el conjunto Resultado del data frame mx.1.csv 

```python
import pandas as pd

# Leer el archivo CSV en un DataFrame de Pandas
df = pd.read_csv('dataliga1/mexico-master/2020s/2020-21/mx.1.csv')

# Crear nuevas columnas con los goles locales y visitantes
df[['goles_local', 'goles_visitante']] = df['FT'].str.split('-', expand=True).astype(int)

# Eliminar la columna "resultado" original
df.drop(['FT'], axis=1, inplace=True)

# Mostrar el resultado
print(df)
```

```python
df
```

```python
df=df.rename(columns={'Team 1': 'Local','Team 2':'Visitante','Date':'Fecha','goles_local':'Goles_local','goles_visitante':'Goles_Visitante'})
```

```python
df
```

```python
df['Local']=df['Local'].replace('Tigres UANL','Tigres')
df['Local']=df['Local'].replace('Querétaro FC','Queretaro')
df['Local']=df['Local'].replace('Santos Laguna','Santos')
df['Visitante']=df['Visitante'].replace('Tigres UANL','Tigres')
df['Visitante']=df['Visitante'].replace('Querétaro FC','Queretaro')
df['Visitante']=df['Visitante'].replace('Santos Laguna','Santos')
```

```python
df['Local']=df['Local'].replace('Deportivo Toluca FC','Toluca')
df['Visitante']=df['Visitante'].replace('Deportivo Toluca FC','Toluca')
```

```python
df[df['Local']=='Toluca']
```

```python
cambios={'CD Guadalajara Chivas':'Chivas','Club León':'Leon','Club Tijuana':'Tijuana','FC Juárez':'Juarez','Atlético San Luis':'Atletico','Club Necaxa':'Necaxa','Pumas UNAM':'Pumas','Club Atlas':'Atlas','Club Puebla':'Puebla','Club América':'America','Mazatlán FC':'Mazatlan','CF Pachuca':'Pachuca','CF Monterrey':'Monterrey'}
df['Local']=df['Local'].replace(cambios)
df['Visitante']=df['Visitante'].replace(cambios)
```

```python
df1=pd.concat([data,df], ignore_index=True)
```

```python
Juarez=df1[df1['Local']=='Juarez']
```

```python
Goles=Juarez.loc[:,'Goles_local']+Juarez.loc[:,'Goles_Visitante']
y = Goles
X=np.array(range(len(Goles))).reshape(-1, 1)
```

```python
y.mean()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regression.predict(X_train), color = "blue")
plt.title("Sueldo vs Años de Experiencia (Conjunto de Entrenamiento)")
plt.xlabel("Semana")
plt.ylabel("Goles")
plt.show()
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_train, regression.predict(X_train), color = "blue")
plt.title("Sueldo vs Años de Experiencia (Conjunto de Testing)")
plt.xlabel("Semana")
plt.ylabel("Goles")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regression = DecisionTreeRegressor(random_state = 0)
regression.fit(X, y)
y_pred = regression.predict([[3]])
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = "red")
plt.plot(X, regression.predict(X), color = "blue")
plt.title("Modelo de Regresión")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```
