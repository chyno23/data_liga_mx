```python
import pandas as pd
```

```python
data=pd.read_html("https://fbref.com/es/comps/31/horario/Resultados-y-partidos-en-Liga-MX")
```

```python
data=data[0]
```

```python
j=0
for i in range(len(data['Marcador'])):
    if data['Marcador'][i]=='Marcador':
        print(i)
        j=j+1
print(j)
```

```python
data['Marcador']=data['Marcador'].str.replace('–','-')
```

```python
data['Marcador'] = data['Marcador'].str.replace(r'\(\d\)', '', regex=True).str.strip()
data['Marcador'] = data['Marcador'].fillna('0-0')
```

```python
data[['Gol_local','Gol_visita']]=data['Marcador'].str.split('-',expand=True).astype(int)
data.drop(['Marcador'],axis=1,inplace=True)
```
