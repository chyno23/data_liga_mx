```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_html("https://fbref.com/es/comps/31/2021-2022/horario/Marcadores-y-partidos-de-2021-2022-Liga-MX",encoding='UTF-8')
data1=pd.read_html("https://fbref.com/es/comps/31/2020-2021/horario/Marcadores-y-partidos-de-2020-2021-Liga-MX",encoding='UTF-8')
```

```python
data=data[0]
data1=data1[0]
```

```python
data['Marcador']=data['Marcador'].str.replace('–','-')
data1['Marcador']=data1['Marcador'].str.replace('–','-')
```

```python
data['Marcador']=data['Marcador'].dropna()
data['Marcador']
```

```python

data['Marcador'] = data['Marcador'].str.replace(r'\(\d\)', '')
data['Marcador'] = data['Marcador'].fillna('0-0')
```

```python
data['Marcador'][200]='0-0'
data[['Gol_local','Gol_visita']]=data['Marcador'].str.split('-',expand=True).astype(int)
data.drop(['Marcador'],axis=1,inplace=True)
data1['Marcador'][200]='0-0'
data1[['Gol_local','Gol_visita']]=data1['Marcador'].str.split('-',expand=True).astype(int)
data1.drop(['Marcador'],axis=1,inplace=True)
```
